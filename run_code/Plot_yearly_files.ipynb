```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

```python
varname = 'SOIL'
year = '1958'
ds1 = xr.open_dataset(f'/data/ycheng/JRA/Data/91-20_climo/{varname}/{varname}.{year}.nc')
ds2 = xr.open_dataset(f'/scratch/ycheng/JRA/Data/91-20_climo_test2/{varname}/{varname}.{year}.nc')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6), subplot_kw={'projection': ccrs.PlateCarree()})
levels = np.linspace(-5, 5, 11)

ds1.anomaly[100].plot(ax=axes[0], transform=ccrs.PlateCarree(),cmap='coolwarm',levels=levels)
axes[0].coastlines()
ds2.anomaly[100].plot(ax=axes[1], transform=ccrs.PlateCarree(),cmap='coolwarm',levels=levels)
axes[1].coastlines()
```
